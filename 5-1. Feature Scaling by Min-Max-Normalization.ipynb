```python
import numpy as np
import pandas as pd

from sklearn import preprocessing
```

```python
NBA_FILE_PATH = 'C:/Users/USER/Desktop/김주호/코딩/Python/2. Machine Learning/NBA_player_of_the_week.csv'
nba_player_of_the_week_df = pd.read_csv(NBA_FILE_PATH)                        #파일 불러오기
```

```python
nba_player_of_the_week_df.head() #데이터 잘 지정 되었는지 확인
```

```python
nba_player_of_the_week_df.describe()   #간단한 통계를 확인해보자
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM','Weight KG','Age']]   # [하고 column지정 위해 또 []넣기]
height_weight_age_df.head()
```

```python
scaler = preprocessing.MinMaxScaler()     # preprocessing의 한 방법인 MinMaxScaler를 설정
```

```python
normalized_data = scaler.fit_transform(height_weight_age_df)   #Normalize 시키기
normalized_data
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Height', 'Weight', 'Age'])
normalized_df.describe()    # 키, 몸무게, 나이를 정규화한 데이터를 표로 만들고 이들의 통계적 수치들을 표현
```
