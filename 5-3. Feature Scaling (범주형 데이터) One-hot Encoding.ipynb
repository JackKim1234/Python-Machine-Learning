```python
import pandas as pd

TITANIC_FILE_PATH = 'C:/Users/USER/Desktop/김주호/코딩/Python/2. Machine Learning/titanic.csv'
```

```python
titanic_df = pd.read_csv(TITANIC_FILE_PATH)
titanic_df.head()
```

```python
titanic_sex_embarked = titanic_df[['Sex','Embarked']]  #필요한 features만 따로 뽑기([행[열]])
```

```python
titanic_sex_embarked
```

```python
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked)   #one-hot 인코딩을 하는 방법
```

```python
one_hot_encoded_df
```

```python
one_hot_encoded_and_other_datas_df = pd.get_dummies(data = titanic_df, columns = ['Sex', 'Embarked']) # 위 결과가 기존 표에 표기
```
